```python
formatos_de_perguntas = {
    'F1': 'Responda qual matéria corresponde à seguinte pergunta? {}. Escolha entre Português, Literatura, Redação, Biologia, Geografia, Matemática, Física, História, Química, Inglês, Espanhol, Filosofia, Sociologia, História da Arte, Orientação de Estudos e Ensino Religioso.',
    'F2': 'Responda qual matéria entre Português, Literatura, Redação, Biologia, Geografia, Matemática, Física, História, Química, Inglês, Espanhol, Filosofia, Sociologia, História da Arte, Orientação de Estudos e Ensino Religioso corresponde à seguinte pergunta? {}.',
    'F3': 'Qual matéria corresponde à seguinte pergunta? {}. Escolha entre Português, Literatura, Redação, Biologia, Geografia, Matemática, Física, História, Química, Inglês, Espanhol, Filosofia, Sociologia, História da Arte, Orientação de Estudos e Ensino Religioso.',
}
```

```python
rows = []
for pergunta, resposta in zip(perguntas, respostas):
    for formato in formatos_de_perguntas:
        pergunta = formatos_de_perguntas[formato].format(pergunta)
        start_time = time.time()
        response = model.generate_content(f"Responda qual matéria corresponde à seguinte pergunta? {pergunta}")
        end_time = time.time()
        row = {
            'formato': formato,
            'tempo(s)': end_time - start_time,
            'resposta': response.text,
            'resposta_correta': resposta,
        }
        rows.append(row)
df = pd.DataFrame(rows)
```
